```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
df_mobile = pd.read_csv('mobile.csv')
df_mobile
```

```python
print("1) - ")
def first_name(x):
    name_list = x.split()
    return name_list[0]
df_mobile['first_name']  = df_mobile['Customer Name'].apply(first_name)
df_mobile
```

```python
print("2) -")
def phone_comp(x):
    digit3 = x.split('-')
    if digit3 == '050':
        digit3[1] == 'pelephone'
    return digit3[0]
df_mobile['company'] = df_mobile['Phone Number'].apply(phone_comp)

# df_mobile['company'].replace('050','pelephone')
# df_mobile['company'].replace('051','golan')
# df_mobile['company'].replace('052','cellcom')
# df_mobile['company'].replace('054','orange')
# df_mobile['company'].replace('053','motorola')

df_mobile['company'] = df_mobile['company'].replace(['050','051','052','053','054'],['pelephone','golan','cellcom','motorola','orange'])
df_mobile
```

```python
display("3-a) ",df_mobile['company'].value_counts())
print()
display("3-b) ",df_mobile['company'].unique())
print()
display("3-c) ",df_mobile['company'].nunique())

```

```python
display("4) -")
df_mobile['user_tape'] = df_mobile.apply(lambda row: 'talker' if row['Talked'] > row['Internet'] else 'surfer',axis=1)
df_mobile
```

```python
display("5) - ")
df_mobile.sort_values(by=['Talked','Internet'], ascending=True, inplace=False)
```

```python
display("6) -")
df_mobile.nlargest(5,'Talked')
```

```python
df_mobile.nsmallest(5,'Internet')
```
